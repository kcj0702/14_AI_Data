```python
# IF문은 왜 필요할까?
# 돈이 있으면 택시를 타고 가고, 돈이 없으면 걸어간다.
# 우유 하나 사와, 달걀 있으면 6개 사와

money = True

if money: print("택시를 타고 간다")
else: print("걸어가라")
```

```python
# 우유 하나 사와, 달걀 있으면 6개 사와

egg = True

if egg: print("우유 6개 사온다")
else: print("우유 1개 사온다")
```

```python
# IF문의 기본 구조

# if 조건문:
# 	수행할문장1
# 	수행할문장2
# 	...
# else:
# 	수행할문장1
# 	수행할문장2
# 	...
```

```python
# x < y 작다
# x > y 크다
# x >= y 크거나 같다
# x <= y 작거나 같다
# x == y 같다
# x != y 같지 않다
```

```python
# 돈이 30000원 이상 있거나 카드가 있다면 택시를 타고 가고,
# 그렇지 않으면 걸어가라

money = 30000
card = True

if (money >= 30000) | (card == True):
	print("택시를 타고 간다")
else:
	print("걸어 간다")
```

```python
# 만약에 주머니에 돈이 있으면 택시를 타고 가고
# 없으면 걸어가라.

pocket = ["paper", "phone", "money"]

if "money" in pocket:
	print("택시를 타고 간다 걷기 싫으니까")
else:
	print("그래도 택시를 타고간다 걷기 싫으니까")
```

```python
# 다양한 조건을 판단하는 elif
# 주머니에 돈이 있으면 택시를 타고가고
# 주머니에 돈이 없지만 카드가 있으면 택시를 타고가고
# 돈도 없고 카드도 없으면 걸어가라

pocket = ["paper", "phone", "money"]
card = True

if "money" in pocket:
	print("택시를 타고간다")
else:
	if card:
		print("택시를 타고간다")
	else:
		print("걸어가지 말고 삼성페이로 택시탄다")
```

```python
# 성적이 90점 이상이면 "A학점"
# 성적이 90점 미만, 80점 이상이면 "B학점"
# 성적이 80점 미만, 70점 이상이면 "C학점"
# 성적이 70점 미만, 60점 이상이면 "D학점"
# 그 외는 "F학점"

score = 65

if score >= 90:
	print("A학점")
elif score >= 80: # score < 90 없어도 됨
	print("B학점")
elif (score < 80) & (score >= 70):
	print("C학점")
elif score >= 60:
	print("D학점")
else:
	print("F학점 낙제")
```

```python
# match-case 구문

# match 변수
# 	case 패턴1:
# 		수행할 문장
# 	case 패턴2:
# 		수행할 문장
```

```python
grade = input("A ~ F 중 하나의 값을 입력하시오")

match grade:
	case "A" | "B" | "C":
		print("합격입니다")
	case _:
		print("불합격입니다")
```
