```python
# While문 기본 구조

# while 조건문:
# 	수행할 문장1
# 	수행할 문장2
# 	수행할 문장3
# 	...
```

```python
treehit = 0
while treehit < 10:
	treehit = treehit + 1
	print(f"나무를 {treehit}번 찍었습니다")
	if treehit == 10:
		print("나무를 베었습니다")
```

```python
number = 0
prompt = """
1. Add
2. Del
3. List
4. Quit
"""

while number != 4:
	print(prompt)
	number = int(input("1 ~ 4까지의 숫자 중 하나를 입력해주세요"))
```

```python
# While 문 강제로 빠져나가기

커피 = 5
돈 = 300
while 돈:
	print("돈을 받았으니 커피를 내립니다.")
	커피 -= 1 # 커피 = 커피 - 1
	print(f"남은 커피는 {커피}잔 입니다.")
	if 커피 == 0:
		print("커피가 매진되었습니다. 장비를 정지합니다.")
		break
print("정지가 안되잖아 안돼ㅐㅐ")
```
