```python
# 파일 생성하기
f = open("새파일.txt", "w") # open("./"): 현재 파일 열기
f.close()
```

```python
f = open("새파일2.txt", "w", encoding="utf-8")
for i in range(1, 11):
	data = f"{i}번째 줄입니다.\n"
	f.write(data)
f.close()
```

```python
f = open("./새파일2.txt", "r", encoding="utf-8")
line = f.readline() # 하나의 줄만 읽음
print(line)
f.close()
```

```python
f = open("./새파일2.txt", "r", encoding="utf-8")
while True:
	line = f.readline()
	if not line: break
	print(line)

f.close()
```

```python
# readlines()
f = open("./새파일2.txt", "r", encoding="utf-8")
lines = f.readlines()
print(lines)
f.close()
```

```python
# 파일에 새로운 내용 추가하기
f = open("./새파일2.txt", "a", encoding="utf-8")
for i in range(11, 21):
	data = f"{i}번째 줄입니다.\n"
	f.write(data)

f.close()
```

```python
# with문과 함께 사용하기
with open("foo.txt", "w") as f: # with문을 쓰면 f.close()를 안해도 됨
	f.write("Life is too short, you need Python")
```
