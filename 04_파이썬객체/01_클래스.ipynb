```python
class Calulator:
	def __init__(self):
		self.result = 0

	def add(self, num):
		self.result += num
		return self.result
	
	def sub(self, num):
		self.result -= num
		return self.result
	
	def mul(self, num):
		self.result *= num
		return self.result
	
	def dvd(self, num):
		if num == 0:
			return "0으로 나눌 수는 없습니다."
		self.result /= num
		return self.result
	
cal1 = Calulator()
print(cal1.add(10))
print(cal1.sub(2))
print(cal1.mul(5))
print(cal1.dvd(4),"\n")

cal2 = Calulator()
print(cal2.add(5))
print(cal2.mul(10))
print(cal2.sub(10))
print(cal2.dvd(5))
```

```python
# 객체에 연산할 숫자 지정하기

class FourCal:
	def setdata(self, first, second):
		self.first = first
		self.second = second

	def add(self):
		result = self.first + self.second
		return result
	def sub(self):
		result = self.first - self.second
		return result
	def mul(self):
		result = self.first * self.second
		return result
	def dvd(self):
		result = self.first / self.second
		return result


cal3 = FourCal()
print(type(cal3))
cal3.setdata(4, 2)

print(cal3.first)
print(cal3.second)
print(cal3.add())
print(cal3.sub())
print(cal3.mul())
print(cal3.dvd())
```

```python
class FourCal:
	def __init__(self, first, second):
		self.first = first
		self.second = second

	def add(self):
		result = self.first + self.second
		return result
	def sub(self):
		result = self.first - self.second
		return result
	def mul(self):
		result = self.first * self.second
		return result
	def dvd(self):
		result = self.first / self.second
		return result
	
cal4 = FourCal(20, 10)
print(cal4.first)
print(cal4.second)
print()
print(cal4.add())
print(cal4.sub())
print(cal4.mul())
print(cal4.dvd())
```

```python
class MoreFourCal(FourCal):
	def pow(self):
		result = self.first ** self.second
		return result
```
