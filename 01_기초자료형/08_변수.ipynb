```python
a = [1, 2, 3]
b = a[:]
print(id(a))
print(id(b))
```

```python
# 변수를 만드는 여러가지 방법
a, b =["python", "like"]
print(a)
print(b)
print(type(a))
a, b = ("python", "like"), ("python", "like")
print(a)
print(b)
print(type(a))
```
