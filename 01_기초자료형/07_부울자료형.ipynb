```python
print(1 == True)
print(2 > 1)
print(2 < 1)
```

```python
print("오재윤은 바보다", bool("오재윤은 바보다"))
print(0, bool(0))
print(2, bool(2))
print("python", bool("python"))
print("", bool(""))
print("[]", bool([]))
print("[1, 2, 3]", bool([1, 2, 3]))
```

```python
if 3 > 1:
	print("참")
else:
	print("거짓")
```

```python
# 논리연산자
# 1. and 연산자
print(True and True)
print(True and False)
print(False and False)
print(True and False and True)

print("-" * 10)

# 2. or 연산자
print(True or True)
print(True or False)
print(False or False)
print(False or False or True)

print("-" * 10)

# 3. not 연산자
print(not True)
print(not False)
print(not 3)
```
