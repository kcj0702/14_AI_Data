```python
# 딕셔너리는 어떻게 만들까?
dic = {"name" : "김수홍무거북이와두루미", "phone" : "010-1234-5678", "birth" : "01-01-01"}
print(type(dic))
```

```python
# 딕셔너리 쌍 추가하기
a = {1 : "a"}
a[2] = "b"
print(a)
a["name"] = "김수홍무거북이와두루미"
print(a)
a[(1,2)] = "응애"
print(a)
a[[3, 4]] = "노수헌"
print(a)
```

```python
# 딕셔너리 요소 삭제하기
a = {1: 'a', 2: 'b', 'name': '김수홍무거북이와두루미', (1, 2): '응애'}
del a[1]
print(a)
del a["name"]
print(a)
```

```python
# 딕셔너리에서 key를 사용해 value 얻기
profile = {
	"name" : "노수헌",
	"age" : (28+26)/2
}
profile["age"]
```

```python
# 딕셔너리에 관련된 함수
# 1. key 리스트 만들기 - keys
print(profile.keys())
profile = {
	"이름" : "노수헌",
	"나이" : (28+26)/2,
	"번호" : "010-1234-5678"
}

# 2. value 리스트 만들기 - values
profile = {
	"이름" : "노수헌",
	"나이" : (28+26)/2,
	"번호" : "010-1234-5678"
}
print(profile.values())

# 3. key:value 쌍 얻기 - items
profile = {
	"이름" : "노수헌",
	"나이" : (28+26)/2,
	"번호" : "010-1234-5678"
}
print(profile.items())

# 4. key, value 쌍 모두 지우기 - clear
profile = {
	"이름" : "노수헌",
	"나이" : (28+26)/2,
	"번호" : "010-1234-5678"
}
print(profile.clear())

# 5. key로 value 얻기 - get
profile = {
	"이름" : "노수헌",
	"나이" : (28+26)/2,
	"번호" : "010-1234-5678"
}
print(profile.get("이름")) # profile["이름"]

# 6. 해당 key가 딕셔너리 안에 있는지 조사하기 - in
profile = {
	"이름" : "노수헌",
	"나이" : (28+26)/2,
	"번호" : "010-1234-5678"
}
print("이름" in profile)

# 7. key로 value 얻기 - pop
profile = {
	"이름" : "노수헌",
	"나이" : (28+26)/2,
	"번호" : "010-1234-5678"
}
temp = profile.pop("이름")
print(temp)
print(profile)
```
