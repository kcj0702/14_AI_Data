```python
# (ctrl + /) 큰 따옴표 둘러싸기
"Life is too short, you need Python"

# 작은 따옴표 둘러싸기
'Python is fun'

# 큰 따옴표 3개를 연속으로 둘러싸기
"""Life is too short, 
you need Python"""

# 작은 따옴표 3개를 연속으로 둘러싸기
'''Life is too short, 
you need Python'''
```

```python
food = "Python's favorite food is perl"
say = '"Python is very easy." he says'

# 역슬래시를 사용해서 작은 따옴표와 큰 따옴표를 문자열에 포함하기
food = 'Python\'s favorite food is perl'
food

say = "\"Python is very easy.\" he says" # (이스케이프 코드)
```

```python
# 줄을 바꾸기 위한 이스케이프 코드 \n
print("Life is too short, \n you need Python")
print("Life is too short, \t you need Python") # tab
```

```python
# 문자열 더하기 연산
a = "Python"
b = " is fun"

a + b
```

```python
# 문자열 곱하기 연산
a = "Python"
a * 2 + b * 2
```
