- DataFrame/Series의 원소/요소의 실제 데이터에 맞게 타입 변환
    * 현재 타입 확인 속성 : dtype/dtypes
    * 현재 타입 변경 메서드 : 변수명.astype(변환_타입명) 반드시 저장 필수

```python
# 모듈 로딩
import pandas as pd             # 데이터 분석용 패키지
import numpy as np              # 배열을 파이썬 언어로 구현한 패키지
                                # pandas에서 데이터 저장을 numpy의 배열로 함
```

```python
# 데이터 준비 및 DF 인스턴스 생성
df = pd.DataFrame({ "이름": ["김민수", "이서연", "박지훈"], 
                   "나이": ["20", "21", np.nan], 
                   "점수": ["85.5", "90.0", "78.5"], 
                   "가입일": ["2024-01-10", "2024-02-15", "2024-03-01"] })
```

```python
# DF 인스턴스 확인
df, df.index, df.columns
```

```python
# DF 인스턴스 요약 정보
df.info()
```

```python
# 실제 데이터 일부
df.head(2)
```

- 방법1) DataFrame/Series.astype(변환타입)

```python
# => 실제 데이터와 타입이 맞지 않는 컬럼 확인
# => 나이 >>> int, 점수 >>> float, 가입일 >>> datetime64
# => 변환 자료형 : 타입명 입력 시 가장 큰 값의 타입 (int => int64, float => float64)
df['나이'] = df['나이'].astype(int)                     # inplace 매개변수 X, 그래서 반드시 저장 
df['점수'] = df['점수'].astype(float)
df['가입일'] = df['가입일'].astype('datetime64[ns]')    # 날짜시간타입명은 반드시 문자열
```

```python

# => 변환 자료형 : 타입명 입력 시 가장 큰 값의 타입 (int => int64, float => float64)
df['나이'] = df['나이'].astype('int8')                     # inplace 매개변수 X, 그래서 반드시 저장 
df['점수'] = df['점수'].astype('float32')
df['가입일'] = df['가입일'].astype('datetime64[ns]')    # 날짜시간타입명은 반드시 문자열
```

```python
# => 변환 자료형 : 타입명 입력 시 가장 큰 값의 타입 (int => int64, float => float64)
# => 결측치 처리 가능한 타입 : 첫 글자 대문자
df['나이'] = df['나이'].astype('Int8')                  # inplace 매개변수 X, 그래서 반드시 저장 
df['점수'] = df['점수'].astype('Float32')
df['가입일'] = df['가입일'].astype('datetime64[ns]')    # 날짜시간타입명은 반드시 문자열 권장 X
```

```python
df.dtypes
```

- 방법2) 숫자형으로 변환 함수 - pandas.to_numeric(데이터)

```python
# 데이터 준비 및 DF 인스턴스 생성
df = pd.DataFrame({ "이름": ["김민수", "이서연", "박지훈"], 
                   "나이": ["20", "21", np.nan], 
                   "점수": ["85.5", "90.0", "78.5"], 
                   "가입일": ["2024-01-10", "2024-02-15", "2024-03-01"] })
```

```python
# 나이, 점수 컬럼 Object ==> 숫자 정수, 실수 형변환
print('변환 전 ====\n', df.dtypes)
```

```python
# 실수 기본 float64 ==> downcast='float' 설정 시 float32
df['점수'] = pd.to_numeric(df['점수'], downcast='float')
```

```python
print('변환 후 ====\n', df.dtypes)
```

- 방법3) 날짜시간 데이터 형변환 - pandas.to_datetime()

```python
print('변환 전 ====\n', df.dtypes)
```

```python
# 가입일 컬럼 object >>> datetime64[ns]
df.가입일 = pd.to_datetime(df.가입일, dayfirst=True)
df.가입일
```

```python
print('변환 후 ====\n', df.dtypes)
```
