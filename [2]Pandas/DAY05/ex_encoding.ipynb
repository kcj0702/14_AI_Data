- 방법1) variety 컬럼 object ==> 숫자/정수 라벨 인코딩

```python
df_map = { "Setosa": 0, "Versicolor": 1, "Virginica": 2 } 

## 결측값 NaN에는 함수 적용하여 NaN 결과를 새로운 컬럼에 추가
df["variety_label"] = df['variety'].map(df_map) 

# 확인
df.head(3)
```

- 방법2) 특정 값만 1로 설정하는 인코딩 : 원-핫 인코딩

```python
df_encoded = pd.get_dummies( df, columns=["variety"], dtype=int ) 
print(df_encoded.tail())
```
