```python
# ============================================================
# 문제 1) 데이터 파일을 읽고 기본 구조를 확인하세요.
# [요구사항]
# 1.  "시도별 전출입 인구수.xlsx" 파일의 "데이터" 시트를 읽어 DataFrame으로 저장하세요.
# 2. 앞부분 5개 행을 출력하세요.
# 3. 데이터의 행수와 열수를 출력하세요.
# 4. 각 컬럼의 자료형을 확인하세요.
# ============================================================
df = pd.read_excel(r'C:\Users\KDT013\14_AI_Data\[2]Pandas\DATA\시도별 전출입 인구수.xlsx', index_col=[0, 1], header=[0, 1])
display(df.head(5))
print(df.shape[0], df.shape[1])
print(df.dtypes)
```

```python
# ============================================================
# 문제 2) 불필요한 행을 삭제하고 인덱스를 초기화하세요.
# [요구사항]
# 1. 데이터 파일을 읽은 후, 두 번째 행에 들어 있는 "이동자수 (명)" 설명 행을 삭제하세요.
# 2. 삭제 후 행 인덱스를 0부터 다시 설정하세요.
# 3. 처리 전과 처리 후의 행과 열 수를 각각 출력하세요.
# 4. 처리 후 앞부분 5개 행을 출력하세요.
# ============================================================
df_af = df.copy()
df_af.columns = [col[0] for col in df.columns]
df_af.reset_index(drop=False, level=[0, 1], names=['전출지별', '전입지별'], inplace=True)
print('처리 전 행:', df.shape[0], '개')
print('처리 전 열:', df.shape[1], '개')
print('처리 후 행:', df_af.shape[0], '개')
print('처리 후 열:', df_af.shape[1], '개')
display(df_af.head())
```

```python
# ============================================================
# 문제 3) 연도 컬럼의 자료형을 숫자형으로 변환하세요.
# [요구사항]
# 1. 1970년부터 2017년까지의 연도 컬럼을 선택하세요.
# 2. 연도별 이동자수 데이터에 포함된 "-" 값을 결측치로 처리할 수 있도록 변환하세요.
# 4. 변환 후 연도 컬럼들의 자료형을 출력하세요.
# 5. 전체 결측치 개수를 출력하세요.
# ============================================================
for col in df_af.columns[2:]:
    df_af[col] = pd.to_numeric(df_af[col], downcast='integer', errors='coerce')
display(df_af.dtypes)
print('전체 결측치 개수', df_af.isna().sum().sum(), '개')
```

```python
# ============================================================
# 문제 4) 특정 행과 열을 선택하세요.
# [요구사항]
# 1. 전출지별, 전입지별, 1970, 1980, 1990, 2000, 2010, 2017 컬럼만 선택하여 새로운 DataFrame을 만드세요.
# 2. 앞에서 10개 행만 출력하세요.
# 3. 전출지별, 전입지별, 2017 컬럼만 출력하세요.
# 4. 0번 행의 2017년 이동자수 값을 선택하여 출력하세요.
# ============================================================
df_new = df_af[['전출지별', '전입지별', 1970, 1980, 1990, 2000, 2010, 2017]]
display(df_new.head(10))
display(df_new[['전출지별', '전입지별', 2017]])
print('2017년 이동자수 값 :', df_new.loc[0, 2017])
```
