#### 【 데이터 병합 】<hr>
- 목적 : 공통 컬럼 값을 기준으로 관련 있는 행끼리 찾아서 연결하는 함수

- 기준 : 특정 컬럼 값

- 사용 : 서로 다른 정보를 공통 기준으로 연결할 경우

- 메서드 : pandas.merge()
    * 주요 매개변수 : on, how

[1] 모듈 로딩 및 데이터 준비 <hr>

```python
## 모듈 로딩
import pandas as pd 
```

```python
## 데이터 파일들
FILE1 = '../DATA/stock price.xlsx'
FILE2 = '../DATA/stock valuation.xlsx'
```

[2] 데이터 로딩 및 기본 정보 확인<hr>

```python
## -----------------------------------
## 데이터 로딩 : 첫번째 행 - 컬럼명 OK
## -----------------------------------
priceDF = pd.read_excel(FILE1)
valuaDF = pd.read_excel(FILE2)
```

```python
## -----------------------------------
## 기본 정보 확인
## -----------------------------------
df_list = [priceDF, valuaDF]

for df in df_list:
    print(f'\n=============')
    df.info()
    print(df.columns)
```

```python
## ==============================================================
## 서로 다른 구조
## stock price.xlsx     컬럼: id, stock_name, value, price
## stock valuation.xlsx 컬럼: id, name, eps, bps, per, pbr
## 공통 컬럼             id
## ==============================================================
```

[2] id 기준으로 두 파일 연결 <hr>

```python
## -----------------------------------
## [2-1] on='id'     : id 값이 양쪽 파일에 모두 있는 데이터만 연결
##       how='inner' : 2개 DF 공통 id 
## -----------------------------------
resultDF = pd.merge(priceDF, valuaDF, on='id', how='inner')

print(f"행 개수     : {resultDF.shape[0] }개")
print(f"열 개수     : {resultDF.shape[1] }개")
print(f"행 인덱스   : {resultDF.index }")
print(f"열 인덱스   : {resultDF.columns }")
display(resultDF)
```

```python
## 동일 내용, 다른 컬럼명 정리 :  stock_name, name 
resultDF = pd.merge(priceDF, valuaDF, on='id', how='inner')
resultDF = resultDF.drop(columns="name")

print(f"행 개수     : {resultDF.shape[0] }개")
print(f"열 개수     : {resultDF.shape[1] }개")
print(f"행 인덱스   : {resultDF.index }")
print(f"열 인덱스   : {resultDF.columns }")
display(resultDF)
```

```python
## -----------------------------------
## [2-2] how ='left' : 왼쪽 파일 기준으로 연결
## -----------------------------------
## => 왼쪽에 행 모두 출력 => 오른쪽 존재하지 않는 값 NaN
resultDF = pd.merge(priceDF, valuaDF, on='id', how='left')

print(f"행 개수     : {resultDF.shape[0] }개")
print(f"열 개수     : {resultDF.shape[1] }개")
print(f"행 인덱스   : {resultDF.index }")
print(f"열 인덱스   : {resultDF.columns }")
display(resultDF)
```

```python
## -----------------------------------
## [2-3] how ='right' : 왼쪽 파일 기준으로 연결
## -----------------------------------
## => 오른쪽에 행 모두 출력 => 왼쪽 존재하지 않는 값 NaN
resultDF = pd.merge(priceDF, valuaDF, on='id', how='right')

print(f"행 개수     : {resultDF.shape[0] }개")
print(f"열 개수     : {resultDF.shape[1] }개")
print(f"행 인덱스   : {resultDF.index }")
print(f"열 인덱스   : {resultDF.columns }")
display(resultDF)
```

```python
## -----------------------------------
## [2-4] how ='outer' : 양쪽 파일 기준으로 연결
## -----------------------------------
## => 오른쪽/왼쪽 행 모두 출력 => 존재하지 않는 값 NaN
resultDF = pd.merge(priceDF, valuaDF, on='id', how='outer')

print(f"행 개수     : {resultDF.shape[0] }개")
print(f"열 개수     : {resultDF.shape[1] }개")
print(f"행 인덱스   : {resultDF.index }")
print(f"열 인덱스   : {resultDF.columns }")
display(resultDF)
```

[3] 특정 컬럼 기준으로 연결<hr>

```python
## -----------------------------------
## [3-1] 특정 컬럼까지 같이 기준으로 연결하고 싶은 경우
## -----------------------------------
## 회사명 컬럼 기준으로 연결
valuaDF = valuaDF.rename(columns={"name": "stock_name"})

resultDF = pd.merge(priceDF, valuaDF, on='stock_name')

print(f"행 개수     : {resultDF.shape[0] }개")
print(f"열 개수     : {resultDF.shape[1] }개")
print(f"행 인덱스   : {resultDF.index }")
print(f"열 인덱스   : {resultDF.columns }")
display(resultDF)
```

```python
## -----------------------------------
## [3-2] 2개 컬럼 기준으로 연결하고 싶은 경우
## -----------------------------------
## id, 회사명 컬럼 기준으로 연결
resultDF = pd.merge(priceDF, valuaDF, on=['id','stock_name'])

print(f"행 개수     : {resultDF.shape[0] }개")
print(f"열 개수     : {resultDF.shape[1] }개")
print(f"행 인덱스   : {resultDF.index }")
print(f"열 인덱스   : {resultDF.columns }")
display(resultDF)
```
