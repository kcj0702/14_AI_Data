#### 【 그룹화 / 그룹연산 】<hr>
- 특정 값을 기준으로 데이터 묶음 후 그룹화
- 그룹별 연산 진행
- 데이터 집계/변환/필터링에 효과적

```python
## 모듈 로딩
import pandas as pd 
```

```python
## DataFrame 인스턴스 생성
df = pd.DataFrame({ '지점': ['서울','서울','서울','부산','부산','대구','대구','대구'],
                    '상품': ['A','A','B','A','B','A','B','B'],
                    '매출': [100,150,200,120,180,90,210,240],
                    '수량': [10,15,20,12,18,9,21,24] })
## 구조 확인
display(df)
```

[1] 그룹화 <hr>

```python
## ----------------------------------------
## 단일 컬럼 그룹화
## ----------------------------------------
## 지점별 그룹화 
g = df.groupby('지점')

## 그룹 속성 확인
print( f"-> 그룹수 : { g.ngroups}개" )           # 그룹 수
print( f"-> 그룹별 : {g.groups}"  )              # 각 그룹의 인덱스
print( f"-> 그룹별 행 수\n{g.size()}" )           # 그룹별 행 수
```

```python
## 그룹별 데이터 추출
for k in g.groups.keys():
    print(f'\n[{k}]=====')
    print(g.get_group(k)) 
```

[2] 그룹연산 - 기본 집계 메서드들 <hr>

```python
## (1) 지점 그룹별 합계
display( g.sum(numeric_only=True))
```

```python
## (2) 지점 그룹별 평균
display( g.mean(numeric_only=True))
```

```python
## (3) 지점 그룹별 여러 집계 동시 적용
g[['매출','수량']].agg(['sum','mean'])
```

```python
## (4) 지점 그룹별 여러 집계 다르게 적용
display( g.agg( {'매출':['sum','mean'], '수량':['max','min']} ) )
```

```python
## (4)지점의 상품별 매출 합계와 평균 
## => 지점별 그룹화 결과에 다시 상품으로 그룹화
for key in g.groups.keys():
    print(f'\n[{key}]=====')
    localDF = g.get_group(key).drop(columns='지점').reset_index(drop=True)
   
    localDF=localDF.groupby('상품')
    display( localDF.agg({'매출':['sum','mean']}) )
```

```python
## ----------------------------------------
## 다중 컬럼 그룹화
## ----------------------------------------

## ==> 지점별 상품에 따른 매출 합계와 평균 집계 
g = df.groupby(['지점','상품'])

## 그룹별 데이터 추출
for k in g.groups.keys():
    print(f'\n[{k}]=====')
    print(g.get_group(k)) 
```

```python
## 지점별 상품에 따른 매출 합계와 평균 집계 
g.agg(['sum','mean'])
```

```python
## -------------------------------------
## agg와 transform 메서드의 차이
## agg()        : 그룹별 결과 출력
## transform () : 행별 결과 출력
## -------------------------------------
g.agg(['sum'])
```

```python
g.transform('sum')
```

[3] agg() / transform() 과 lambda <hr>

```python
## DataFrame 인스턴스 생성
df = pd.DataFrame({ '지점': ['서울','서울','서울','부산','부산','부산','대구','대구','대구'],
                    '매출': [100,150,200,120,180,160,90,210,240] })
## 확인
display(df)
```

```python
## ------------------------------------------
## (1) 그룹별 평균 매출 대비 비율
## ------------------------------------------
df['매출비율'] = df.groupby('지점')['매출'].transform(lambda x: x / x.mean())

## 같은 지점 내에서 매출이 
## 평균보다 높은지/낮은지 한눈에 파악
display(df)

## 지점별로 매출의 범위(최대값 - 최소값)를 계산
# lambda x에서 x는 각 지점별 '매출' 시리즈(Series)를 의미
branch_range = df.groupby('지점')['매출'].agg(lambda x: x.max() - x.min())
print("지점별 매출 변동 폭:")
display(branch_range)
```

```python
df['표준화'] = df.groupby('지점')['매출'].transform(lambda x: (x - x.mean()) / x.std())
display(df)
```

```python
df['순위'] = df.groupby('지점')['매출'].transform(lambda x: x.rank(ascending=False))
display(df)
```
