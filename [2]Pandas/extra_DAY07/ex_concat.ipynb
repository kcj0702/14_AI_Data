#### 【 데이터 연결 】<hr>
- 목적 : 데이터를 방향만 정해서 단순히 이어붙이는 함수

- 기준 : 행 인덱스 또는 컬럼 방향
- 사용 : 같은 구조 즉, 동일 컬럼의 파일을 여러 개 합칠 때 사용
    * 컬럼이 동일하지 않아도 연결은 가능하지만
    * 좋은 concat 데이터는 컬럼 구조가 동일한 데이터

- 메서드 : pandas.concat()
    * 주요 매개변수 : axis, join


[1] 모듈 로딩 및 데이터 준비 <hr>

```python
## 모듈 로딩
import pandas as pd 
```

```python
## 데이터 파일들
FILE1 = '../DATA/stock_price_2024_01.xlsx'
FILE2 = '../DATA/stock_price_2024_02.xlsx'
FILE3 = '../DATA/stock_price_2024_03.xlsx'
FILE4 = '../DATA/stock_price_2024_04.xlsx'
```

[2] 데이터 로딩 및 기본 정보 확인<hr>

```python
## -----------------------------------
## 데이터 로딩 : 첫번째 행 - 컬럼명 OK
## -----------------------------------
janDF = pd.read_excel(FILE1)
febDF = pd.read_excel(FILE2)
marDF = pd.read_excel(FILE3)
apiDF = pd.read_excel(FILE4)
```

```python
## -----------------------------------
## 기본 정보 확인
## -----------------------------------
df_list = [janDF, febDF, marDF, apiDF]

for df in df_list:
    print(f'\n=============')
    df.info()
    print(df.columns)
```

```python
## ==============================================================
## 동일한 컬럼 구조 
## 컬럼 : 'date', 'id', 'stock_name', 'market', 'price', 'volume'
## 3개 DF 연결
## ==============================================================
```

[2] DataFrame 연결 - 동일 인덱스의 행 <hr>

```python
## -----------------------------------
## [2-1] axis=0 : 위아래 방향으로 연결
## -----------------------------------
resultDF = pd.concat( [janDF, febDF, marDF], 
                      axis=0 )

print(f"행 개수     : {resultDF.shape[0] }개")
print(f"열 개수     : {resultDF.shape[1] }개")
print(f"행 인덱스   : {resultDF.index }")
print(f"열 인덱스   : {resultDF.columns }")
display(resultDF)
```

```python
## 3개 DF의 행 인덱스 정리 필요 : ignore_index=True
resultDF = pd.concat(  [janDF, febDF, marDF], 
                       axis=0,
                       ignore_index=True )

print(f"행 개수     : {resultDF.shape[0] }개")
print(f"열 개수     : {resultDF.shape[1] }개")
print(f"행 인덱스   : {resultDF.index }")
print(f"열 인덱스   : {resultDF.columns }")
display(resultDF)
```

```python
## keys 사용하여 중복 행 인덱스 식별 
## ==> MultiIndex Index 형태가 됨!
resultDF = pd.concat( [janDF, febDF, marDF], 
                      axis=0,
                      keys=["2024_01", "2024_02", "2024_03"] )

print(f"행 개수     : {resultDF.shape[0] }개")
print(f"열 개수     : {resultDF.shape[1] }개")
print(f"행 인덱스   : {resultDF.index }")
print(f"열 인덱스   : {resultDF.columns }")
display(resultDF)
```

```python
## -----------------------------------
## [2-2] axis=1 : 좌우 방향으로 연결
## -----------------------------------
## 컬럼명 중복 발생!!!
resultDF = pd.concat( [janDF, febDF, marDF], 
                      axis=1 )

print(f"행 개수     : {resultDF.shape[0] }개")
print(f"열 개수     : {resultDF.shape[1] }개")
print(f"행 인덱스   : {resultDF.index }")
print(f"열 인덱스   : {resultDF.columns }")
display(resultDF)
```

```python
## keys : 사용하면 어느 파일에서 온 컬럼인지 구분
## ==> MultiIndex Column 형태가 됨!
resultDF = pd.concat( [janDF, febDF, marDF], 
                      axis=1,
                      keys=["2024_01", "2024_02", "2024_03"] )

print(f"행 개수     : {resultDF.shape[0] }개")
print(f"열 개수     : {resultDF.shape[1] }개")
print(f"행 인덱스   : {resultDF.index }")
print(f"열 인덱스   : {resultDF.columns }")
display(resultDF)
```

[3] 행 인덱스 다른 DataFrame 연결 <hr>

```python
## -----------------------------------
## [3-1] axis=0 : 위아래 방향으로 연결. 행 인덱스 다름
## -----------------------------------
resultDF = pd.concat( [janDF, apiDF], 
                      axis=0 )

print(f"행 개수     : {resultDF.shape[0] }개")
print(f"열 개수     : {resultDF.shape[1] }개")
print(f"행 인덱스   : {resultDF.index }")
print(f"열 인덱스   : {resultDF.columns }")
display(resultDF)
```

```python
## -----------------------------------
## [3-2] axis=1 : 좌우 방향으로 연결. 행 인덱스 다름 => NAN발생
## -----------------------------------
## 모든 행 연결 : join='outer' [기]
resultDF = pd.concat( [janDF, apiDF], 
                      axis=1 )

print(f"행 개수     : {resultDF.shape[0] }개")
print(f"열 개수     : {resultDF.shape[1] }개")
print(f"행 인덱스   : {resultDF.index }")
print(f"열 인덱스   : {resultDF.columns }")
display(resultDF)
```

```python
## -----------------------------------
## [3-3] axis=1 : 좌우 방향으로 연결. 행 인덱스 다름 => NAN발생
## -----------------------------------
## 모두 존재하는 행 인덱스만 연결 : join='inner'
resultDF = pd.concat( [janDF, apiDF], 
                      axis=1,
                      join='inner' )

print(f"행 개수     : {resultDF.shape[0] }개")
print(f"열 개수     : {resultDF.shape[1] }개")
print(f"행 인덱스   : {resultDF.index }")
print(f"열 인덱스   : {resultDF.columns }")
display(resultDF)
```
